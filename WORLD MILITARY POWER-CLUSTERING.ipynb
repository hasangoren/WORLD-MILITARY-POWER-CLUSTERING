```python
import pandas as pd
import re


def clean_and_convert(value):
    if isinstance(value, str):
        # Harfleri kaldırma
        value = re.sub(r'[^\d,.-]', '', value)
        # Virgülleri noktalarla değiştirme
        value = value.replace(',', '.')
        # Sayıya dönüştürme
        try:
            return float(value)
        except ValueError:
            return '0'
    return value
```

```python
df = df.applymap(clean_and_convert)
```

```python
df= df.applymap(lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)
```

```python
#wcss= within cluster sum of squares

wcss=[]

ss=[]

for i in range(2,10):

    model=KMeans(i)

    model=model.fit(x)

    tahmin=model.predict(x)

    ss1=silhouette_score(x,tahmin)

    ss.append(ss1)

    print(ss1)

    wcss.append(model.inertia_)
    
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```
